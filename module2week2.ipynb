```python
#module 2 week 2 quiz

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "genre": ["Pop", "Rock", "Hip-Hop", "Electronic"],
    "streams": [120, 90, 150, 70]
}

df = pd.DataFrame(data)

# Write code below to create and display a bar plot showing streams by genre
plt.bar(df['genre'], df['streams'],color='skyblue', edgecolor='navy')
plt.xlabel('Genre')
plt.ylabel('Streams')
plt.title('Streams by Genre')
```
