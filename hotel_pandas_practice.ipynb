```python
import pandas as pd
```

```python
df = pd.read_csv("hotel_bookings.csv")
print(df.head())
```

```python
#drop/rename columns
df.columns
```

```python
df.rename({'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, axis=1, inplace=True)
df.head(10)
```

```python
#NaNs percentage
df.isnull().sum() *100 / len(df)
```

```python
#unique values in agent to see if remove
df['agent'].unique()
```

```python
#replace null/nan values with filler value and delete null rows on num_children column
df['agent'].fillna(-1, inplace=True)
df[df['num_children'].isna()]
df[df['country'].isna()]
```

```python

df.dropna(subset=['num_children'], inplace=True)
```

```python
#drop company column due to 94% null
df.drop(columns=['company'], inplace=True)
```

```python
#run nan percentage again
df.isnull().sum() *100 / len(df)
```

```python
df['country'].fillna('unknown', inplace=True)
df.fillna({'country': 'unknown'}, inplace=True)
```

```python
df.fillna({'agent':-1}, inplace=True)
```

```python
#run nan percentage again
df.isnull().sum() *100 / len(df)
```

```python
#check column data types
df.dtypes
```

```python
df = df.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int', 'agent': 'int', })
df.dtypes
```

```python
#bin columns
df['lead_time'].unique()
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']

df['lead_time_binned'] = pd.cut(df['lead_time'], bins = bins, labels=labels)
df[['lead_time', 'lead_time_binned']]
```

```python
#string cleaning
df['hotel'].unique()
```

```python
#remove duplicates
df.loc[df.duplicated(keep=False)]

df.drop_duplicates(keep="first", inplace = True)

df.loc[df.duplicated(keep=False)]
```
